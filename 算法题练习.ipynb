#### LC384. Shuffle an array
Shuffle a set of numbers without duplicates.


```python
import random
def shuffle(array):
    for i in xrange(len(array)):
        swap_index = random.randrange(i,len(array))
        array[i],array[swap_index] = array[swap_index],array[i]
    return array
```

```python
array = [1,2,3,4,5]
shuffle(array)
```

#### LC387. First Unique Character in a String
Given a string, find the first non-repeating character in it and return it's index. If it doesn't exist, return -1.



```python
def firstUniqChar(s):
    for i in range(len(s)):
        c = s[i]
        if s.count(c)==1:
            return i
    return -1
```

```python
firstUniqChar('zxsvefzx')
```

```python
# 选取第一个重复出现的字母 从一个string 中
def firstRecurChar(s):
    for i in range(len(s)):
        c = s[i]
        if s.count(c)!=1:
            return i
    return -1    
```

```python
firstRecurChar('Aafinding')
```

#### deduplicate and keep order

```python
def unique(nums):
    unique = []
    for n in nums:
        if n not in unique:
            unique.append(n)
    return unique
```

#### find number of the substrings in an input string

```python
def find_sub(string):
    a = set()
    for i in range(len(string)):
        j = 1
        while True:
            for i in xrange(len(string)-j+1):   
                a.add(string[i:i+j])
            if j==len(string):   
                break
            j+=1
        return sorted(list(a))
```

```python
def permute(num):
    if len(num) == 0: 
        return []
    if len(num) == 1: 
        return [num]
    res = []
    for i in range(len(num)):
        for j in permute(num[:i] + num[i+1:]):
            res.append([num[i]] + j)
    return res
```

```python
def permutation(lst): 
  
    # If lst is empty then there are no permutations 
    if len(lst) == 0: 
        return [] 
  
    # If there is only one element in lst then, only 
    # one permuatation is possible 
    if len(lst) == 1: 
        return [lst] 
  
    # Find the permutations for lst if there are 
    # more than 1 characters 
  
    l = [] # empty list that will store current permutation 
  
    # Iterate the input(lst) and calculate the permutation 
    for i in range(len(lst)): 
        m = lst[i] 
  
       # Extract lst[i] or m from the list.  remLst is 
       # remaining list 
        remLst = lst[:i] + lst[i+1:] 
  
       # Generating all permutations where m is first 
       # element 
        for p in permutation(remLst): 
            l.append([m] + p) 
    return l 
```

#### Two sum
<p>Given an array of integers, return indices of the two numbers such that they add up to a specific target. 
You may assume that each input would have exactly one solution.

<p>Example: 
<p>Given nums = [2, 7, 11, 15], target = 9, Because nums[0] + nums[1] = 2 + 7 = 9, return [0, 1].


```python
def twosum(nums, target):
    # create an dictionary
    d = {}
    for i in xrange(len(nums)):
        x = nums[i]
        if target - x in d:
            return (d[target-x]+1, i+1)
        d[x]=i
```

```python
twosum([1,4,7,10,13], 11)
```

```python
#17
def MaxProfit(p):
    if len(p)<=1: 
        return 0 
    max_profit, min_price = 0, p[0] 
    for i in range(1, len(p)):
        min_price = min(min_price, p[i])
        max_profit = max(max_profit, p[i]-min_price) 
    return max_profit
```
